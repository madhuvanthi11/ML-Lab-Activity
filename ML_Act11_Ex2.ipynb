```python
#importing the required libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#sample dataset

data = {
    'Climate':        [0.190, 0.017, 0.207],
    'Housing':        [0.544, 0.020, 0.204],
    'Health':         [0.782, -0.605, 0.144],
    'Crime':          [0.365, 0.294, 0.585],
    'Transportation': [0.585, 0.085, 0.234],
    'Education':      [0.394, -0.273, 0.027],
    'Arts':           [0.985, 0.126, -0.111],
    'Recreation':     [0.520, 0.402, 0.519],
    'Economy':        [0.142, 0.150, 0.239]
}

df=pd.DataFrame(data, index=['1','2','3'])
print(df)
```

```python
#identifying dominant variables per pricipal component

# Find the absolute value of loadings
abs_df = df.abs()

# Find the variable with highest loading for each PC
dominant_variables = abs_df.idxmax()
print("\nMost dominant variable in each Principal Component:")
print(dominant_variables)
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='coolwarm')
plt.title("PCA Component Loadings")
plt.show()
```
