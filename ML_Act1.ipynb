```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Salary_Data.csv')
```

```python
#Exploring dataset
data.head()
```

```python
data.shape
```

```python
#Checking for any null values in dataset

data.isnull().sum()
```

```python
#Assigning Dependent and Independent variables

x=data.iloc[:,:1].values
y=data.iloc[:, 1:2].values
```

```python
#Splitting the dataset into Training and Testing Dataset

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Fitting the Model (Linear Regression)

model=LinearRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)

print(y_pred)
print(y_test)
```

```python
#Plot for Training dataset

plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('SALARY vs EXPERIENCE (training set)')
plt.xlabel('Experience in Years')
plt.ylabel('Salary in Rupees')
plt.show()
```

```python
#Plot for Testing dataset

plt.scatter(x_test, y_test, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('SALARY vs EXPERIENCE (testing set)')
plt.xlabel('Experience in Years')
plt.ylabel('Salary in Rupees')
plt.show()
```

```python
#Train a model to predict the fare of a taxi ride in Chicago, Illoinis

df=pd.read_csv('taxi.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.shape
```

```python
df=df[df['Trip Miles']>0]
df=df[df['Fare']>0]
```

```python
x=df[['Trip Miles']]
y=df['Fare']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model=LinearRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
```

```python
print(y_pred)
print(y_test)
```

```python
plt.scatter(x_train, y_train, color='purple', edgecolor='black')
plt.plot(x_train, model.predict(x_train), color='green')
plt.title('FARE vs TRIP MILES (training set)')
plt.xlabel('Trip Miles')
plt.ylabel('Fare')
plt.show()
```

```python
#In the above graph, most of the data points are clustered around the corner because the dataset predominantly has short distance trips. We can limit the axes to zoom in the plot
plt.scatter(x_train, y_train, color='purple', edgecolor='black')
plt.plot(x_train, model.predict(x_train), color='green')
plt.title('FARE vs TRIP MILES (training set)')
plt.xlabel('Trip Miles')
plt.ylabel('Fare')
plt.xlim(0, 50)
plt.ylim(0, 200)
plt.show()
```

```python
#To show the distribution of trip miles (consists mostly of short trips)
df['Trip Miles'].hist(bins=50)
plt.title("Distribution of Trip Miles")
plt.xlabel("Trip Miles")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(x_test, y_test, color='purple', edgecolor='black')
plt.plot(x_train, model.predict(x_train), color='green')
plt.title('FARE vs TRIP MILES (testing set)')
plt.xlabel('Trip Miles')
plt.ylabel('Fare')
plt.show()
```
